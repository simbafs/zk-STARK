# Finite Field

```python
M = 17
pr = []
for a in range(1, M):
    for i in range(1, M):
        if a**i % M == 1:
            if i == M-1:
                pr.append(a)
            print(f'{a}:\t{i}')
            break
    else:
        print(f'can not find primitive root of {a}')

print(pr)
```

```python
N = 4
gs = set()
for omega in pr:
    g = (omega ** int((M-1)/N)) % M # omage ** 4
    gs.add(g)
    print(f'{omega = },\t{g = }')
```

```python
int((M-1)/N)
```

```python
for g in gs:
    G = set([g**i % M for i in range(M)]) # actually range N is enough
    print(f'when {g = }, \t {G = }')
    assert len(G) == N
```

```python
rho = 2
L_size = N*rho # 8
h = pr[0]**int((M-1)/L_size) % M # omega ** 2
print(f'{h = }')
L = [h**i % M for i in range(N*rho)]
print(f'{L = }')
```

# VM

```python
instructions = [
    'set 1',
    'swap',
    'set 2',
    'add',
    'set 4',
    'mul',
]

class VM(object):
    def __init__(self):
        self.r1 = 0
        self.r2 = 0
        self.N = 0
        self.trace = []

    def exec(self, instructions: [str]):
        self.N = len(instructions)

        for inst in instructions:
            args = inst.split(' ')
            self.eval(args)


        self.trace.append([self.r1, self.r2, 0, 0, 0, 0, 0, 0])
        self.N = len(self.trace)

        return self.trace

    def eval(self, args: [str]):
        op = args[0]
        args = args[1:]

        if op == 'set':
            if len(args) < 1:
                raise  Exception(f'op set should has one arg')
    
            self.trace.append([self.r1, self.r2, int(args[0]), 1, 0, 0, 0, 0])
            self.r2 = int(args[0])
        elif op == 'swap':
            self.trace.append([self.r1, self.r2, 0, 0, 1, 0, 0, 0])
            self.r1, self.r2 = self.r2, self.r1
        elif op == 'add':
            self.trace.append([self.r1, self.r2, 0, 0, 0, 1, 0, 0])
            self.r1 = self.r1 + self.r2
        elif op == 'mul':
            self.trace.append([self.r1, self.r2, 0, 0, 0, 0, 1, 0])
            self.r1 = self.r1 * self.r2
        elif op == 'nop':
            self.trace.append([self.r1, self.r2, 0, 0, 0, 0, 0, 1])
        else:
            raise Exception(f'Unknown opcode: {op}')


vm = VM()
vm.exec(instructions)
```
